```python
import pygame
import time

pygame.init()


def get(i,j):
    x = 5 #edge_width
    y = 5 #edge_width
    x += 65*i #rect_side+edge_width
    y += 65*j #rect_side+edge_width
    x += (60-(30-10))//2 #getting to centre of the cell
    y += (60-(30-10))//2 #getting to centre of the cell
    return x,y

#start of sudokusolving logic
class Sudoku:
    def __init__(self,sudoku):
        size = len(sudoku)
        self.sudoku = sudoku
        self.row = [[False for x in range(size)] for y in range(size)]
        self.col = [[False for x in range(size)] for y in range(size)]
        self.box = [[[False for x in range(size)] for y in range(size//3)] for z in range(size//3)]
        for i in range(size):
            for j in range(size):
                if sudoku[i][j]>0:
                    self.row[i][sudoku[i][j]-1]=True
                    self.col[j][sudoku[i][j]-1]=True
                    self.box[i//3][j//3][sudoku[i][j]-1]=True

def next(i,j,size):
    if j==(size-1):
        return i+1,0
    return i,j+1

def check_possible(obj,i,j,val):
    if obj.row[i][val-1]:
        return False
    if obj.col[j][val-1]:
        return False
    temp_i,temp_j = i//3,j//3
    if obj.box[temp_i][temp_j][val-1]:
        return False
    return True

def fill(obj,i,j,val,screen,nums):
    obj.sudoku[i][j] = val
    
    x,y = get(i,j)
    screen.blit(nums[val],(x,y))
    update_rect = pygame.Rect(x,y,30,30)
    pygame.display.update(update_rect)
    #pygame.time.delay(1000)
    #time.sleep(0.1)
    #pygame.display.flip()
    
    obj.row[i][val-1] = True
    obj.col[j][val-1] = True
    obj.box[i//3][j//3][val-1] = True
    
def defill(obj,i,j,screen,nums):
    val = obj.sudoku[i][j]
    
    x,y = get(i,j)
    screen.blit(nums[val],(x,y))
    update_rect = pygame.Rect(x,y,30,30)
    pygame.display.update(update_rect)
    #pygame.time.delay(1000)
    #pygame.display.flip()
    
    obj.sudoku[i][j] = 0
    obj.row[i][val-1] = False
    obj.col[j][val-1] = False
    obj.box[i//3][j//3][val-1] = False

def sudoku_solver(obj,i,j,size,screen,nums):
    if i>=size or j>=size:
        return True
    if obj.sudoku[i][j]>0:
        i,j = next(i,j,size)
        return sudoku_solver(obj,i,j,size,screen,nums)
    for val in range(1,10):
        if check_possible(obj,i,j,val):
            fill(obj,i,j,val,screen,nums)
            temp_i,temp_j = next(i,j,size)
            if sudoku_solver(obj,temp_i,temp_j,size,screen,nums):
                return True
            defill(obj,i,j,screen,nums)
    return False

sudoku = [
        [3,9,2,1,5,4,0,0,8],
        [8,0,4,0,9,3,0,1,0],
        [7,5,0,2,0,8,0,3,0],
        [0,7,0,0,2,0,8,6,0],
        [2,0,0,0,8,0,0,0,1],
        [0,8,9,0,7,0,0,4,0],
        [0,3,0,5,0,2,0,8,6],
        [0,1,0,9,3,0,5,0,7],
        [6,0,0,8,1,7,4,9,3]
    ]

obj = Sudoku(sudoku)
















#start of pygame logic

def next_pixel(x,y,i,j,width,rect_side,edge_width):
    temp_y=y+rect_side+edge_width
    if temp_y<width:
        return x,temp_y,i,j+1
    return x+rect_side+edge_width,edge_width,i+1,0

def centre(x,rect_side,num_size):
    return x+(rect_side-num_size)//2


rect_side = 60
edge_width = 5
num_size = 30
width,height = 2*[9*rect_side+10*edge_width]
white = (255,255,255)
black = (0,0,0)


screen = pygame.display.set_mode((width,height))
pygame.display.set_caption('Sudoku Solver')
font = pygame.font.Font('freesansbold.ttf', num_size)
run = True
screen.fill(black)
nums = [font.render(str(x),True,black,white) for x in range(10)]
nums[0] = font.render('_',True,black,white)

#creating the cells
def fillcells(obj,width,rect_side,edge_width,nums,screen):
    x,y,i,j,index = edge_width,edge_width,0,0,0
    while x<width and y<height:
        pygame.draw.rect(screen,white,pygame.Rect(x,y,rect_side,rect_side))
        #main command for fillinf the numbers
        screen.blit(nums[obj.sudoku[index//9][index%9]],(centre(x,rect_side,num_size-10),centre(y,rect_side,num_size-10)))
    
        #if i%3==2:
        #x+=edge_width
        #if j%3==2:
        #y+=edge_width
        x,y,i,j = next_pixel(x,y,i,j,width,rect_side,edge_width)
        index=index+1

fillcells(obj,width,rect_side,edge_width,nums,screen)    
pygame.display.flip()
sudoku_solver(obj,0,0,9,screen,nums)
#time.sleep(5)
while run:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
        #if event.type == pygame.KEYDOWN:
        #    fillcells(obj,width,rect_side,edge_width,nums,screen)
    #pygame.display.flip()
            
            
pygame.quit()
print(sudoku)
```
