```python
class Sudoku:
    def __init__(self,sudoku):
        size = len(sudoku)
        self.sudoku = sudoku
        self.row = [[False for x in range(size)] for y in range(size)]
        self.col = [[False for x in range(size)] for y in range(size)]
        self.box = [[[False for x in range(size)] for y in range(size//3)] for z in range(size//3)]
        for i in range(size):
            for j in range(size):
                if sudoku[i][j]>0:
                    self.row[i][sudoku[i][j]-1]=True
                    self.col[j][sudoku[i][j]-1]=True
                    self.box[i//3][j//3][sudoku[i][j]-1]=True
                

def next(i,j,size):
    if j==(size-1):
        return i+1,0
    return i,j+1

def check_possible(obj,i,j,val):
    if obj.row[i][val-1]:
        return False
    if obj.col[j][val-1]:
        return False
    temp_i,temp_j = i//3,j//3
    if obj.box[temp_i][temp_j][val-1]:
        return False
    return True

def fill(obj,i,j,val):
    obj.sudoku[i][j] = val
    obj.row[i][val-1] = True
    obj.col[j][val-1] = True
    obj.box[i//3][j//3][val-1] = True
    
def defill(obj,i,j):
    val = obj.sudoku[i][j]
    obj.sudoku[i][j] = 0
    obj.row[i][val-1] = False
    obj.col[j][val-1] = False
    obj.box[i//3][j//3][val-1] = False

def sudoku_solver(obj,i,j,size):
    if i>=size or j>=size:
        return True
    if obj.sudoku[i][j]>0:
        i,j = next(i,j,size)
        return sudoku_solver(obj,i,j,size)
    for val in range(1,10):
        if check_possible(obj,i,j,val):
            fill(obj,i,j,val)
            temp_i,temp_j = next(i,j,size)
            if sudoku_solver(obj,temp_i,temp_j,size):
                return True
            defill(obj,i,j)
    return False
    
def print_sudoku(sudoku):
    size = len(sudoku)
    for i in range(size):
        for j in range(size):
            if sudoku[i][j]<1:
                print('_',end = ' ')
            else:
                print(sudoku[i][j],end = ' ')
        print()

def get_sudoku(size):
    print('write your starting sudoku with empty cells with 0 :')
    sudoku = []
    for i in range(size):
        sudoku.append(list(map(int,input().split(' '))))
    return sudoku


#sudoku = [[0 for x in range(size)] for y in range(size)]
def main_function():
    #size = int(input('What is the size of the sudoku you want : '))
    #sudoku = get_sudoku(size)
    sudoku = [
        [3,9,2,1,5,4,0,0,8],
        [8,0,4,0,9,3,0,1,0],
        [7,5,0,2,0,8,0,3,0],
        [0,7,0,0,2,0,8,6,0],
        [2,0,0,0,8,0,0,0,1],
        [0,8,9,0,7,0,0,4,0],
        [0,3,0,5,0,2,0,8,6],
        [0,1,0,9,3,0,5,0,7],
        [6,0,0,8,1,7,4,9,3]
    ]
    obj = Sudoku(sudoku)
    temp = sudoku_solver(obj,0,0,len(sudoku))
    print_sudoku(sudoku)
    print('\n\n')
    if temp:
        print('Hurrah we have solved the sudoku')        
    else:
        print('It was not possible to solve it.')

main_function()
```
